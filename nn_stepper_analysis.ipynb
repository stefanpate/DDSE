```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Feedforward performance on lorenz prediction

save_dir = '/cnl/data/spate/Corn/'
ml, mh = 20, 24
last_t = np.loadtxt(save_dir + f"ff_nn_predict_lorenz_test_last_t_models_{ml}_{mh}.csv", delimiter=',')
last_t_mean, last_t_std = np.mean(last_t, axis=1), np.std(last_t, axis=1)
model_numbers = np.arange(ml, mh + 1)
batch_size = last_t.shape[1]
w = 0.5 # Bar width

fig, ax = plt.subplots(figsize=(12,6))

ax.bar(model_numbers, last_t_mean, width=w, yerr=last_t_std, color=(0,0,0,0), edgecolor=(0,0,0,1))

for i, elt in enumerate(model_numbers):
    ax.scatter(elt + np.random.rand(batch_size) * w - w / 2, last_t[i,:], color=(0,0,0,0.5))

ax.set_xlabel('model #')
ax.set_ylabel('time')

plt.show()
```
