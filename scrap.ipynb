```python
theta = np.linspace(0, 2 * np.pi)
a = 1
b = 1
r = a + b * np.sin(theta)

fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(theta, r)
plt.show()
```

```python
r = np.sqrt((a**2) * np.cos(2 * theta))
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(theta, r)
plt.show()
```

```python
r = a * theta
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(theta, r)
plt.show()
```

```python
n = 4
r = a * np.sin(n * theta)
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(theta, r)
plt.show()
```
